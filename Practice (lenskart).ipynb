```python
import pandas as pd
import bs4
import requests
```

```python
url='https://www.lenskart.com/eyeglasses/collections/john-jacobs-dreamscapes.html'
```

```python
req=requests.get(url)
req
```

```python
soup=bs4.BeautifulSoup(req.text,'html.parser')
soup
print(soup.prettify())
```

```python
name=soup.find_all('div',class_='product-name')
print(name)
```

```python
price=soup.find_all('span',class_='text-color-dark-blue fw700')
print(price)
```

```python
rating=soup.find_all('span',class_='rating-value')
print(rating)
```

```python
description=soup.find_all('div',class_='font-family-regular-new-plp')
print(description)
```

```python
discount=soup.find_all('span',class_='text-color-light-grey')
print(discount)
```

```python
user_r=soup.find_all('span',class_='rating-count')
print(user_r)
```

```python
nm=[]
pr=[]
rt=[]
ds=[]
dis=[]
ur=[]

for i in name:
    nm.append(i.text)
    
for i in price:
    pr.append(i.text)
    
for i in rating:
    rt.append(i.text)
    
for i in description:
    ds.append(i.text)
    
for i in discount:
    dis.append(i.text)
    
for i in user_r:
    ur.append(i.text)
```

```python
print(len(nm))
```

```python
data={"Name":nm,"description":ds,"price":pr,"User Voting":ur,"Rating":rt,"Real Price":dis}
df=pd.DataFrame(data)
```
